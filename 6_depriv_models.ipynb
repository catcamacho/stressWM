```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

project_home = '/data/perlman/moochie/user_data/CamachoCat/combinedDWI'

# read in depriv_data
depriv_data = pd.read_csv('/data/perlman/moochie/user_data/ChinJenna/StressNeurodev/scored_data/depriv_20201216.csv')

ADIs = ['ADI_state_decile', 'ADI_national_percentile']  

# create column reverse score income bins
depriv_data['h_income_bin_reverse'] = 9 - depriv_data['h_income_bin']

# create column reverse score parent_ed bins
depriv_data['parent_ed_bin_reverse'] = 9 - depriv_data['parent_ed_bin']


# standardize ADIs, reversed scored income bin, parent ed 
sc = StandardScaler()
for x in ADIs + ['h_income_bin_reverse'] + ['parent_ed_bin_reverse']:
    depriv_data[x +'_std'] = sc.fit_transform(depriv_data[x].to_numpy().reshape(-1,1))

# create depriv_comp (sum standardized scores)
# depriv_data['depriv_comp'] = depriv_data['ADI_state_decile_std'] +depriv_data['ADI_national_percentile_std'] + depriv_data['h_income_bin_reverse_std'] + depriv_data['parent_ed_bin_reverse_std']

```

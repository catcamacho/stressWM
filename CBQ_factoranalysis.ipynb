```python
import pandas as pd

data = pd.read_csv('../scored_data/affect_scores_combined_20201217.csv',index_col=0)
data.columns
```

```python
cbqdata = data.drop(labels=['map_temploss'], axis=1)
cbqdata
```

```python
from factor_analyzer import FactorAnalyzer

fa = FactorAnalyzer(rotation='varimax', impute='drop', n_factors=3)
fa.fit(cbqdata)
print(fa.get_factor_variance())
loadings = pd.DataFrame(fa.loadings_,columns=['f1','f2','f3'], index=cbqdata.columns)
loadings=loadings.sort_values('f1')
loadings
```

```python
data['cbqsf_NEG'] = cbqdata[['cbqsf_angfrus','cbqsf_discomf','cbqsf_sad','cbqsf_fear','cbqsf_shy']].mean(axis=1)
data['cbqsf_EC'] = cbqdata[['cbqsf_lip','cbqsf_smilelaugh','cbqsf_percepsens','cbqsf_inhibctl',
                            'cbqsf_fallreact','cbqsf_attenfoc']].mean(axis=1)
data['cbqsf_SUR'] = cbqdata[['cbqsf_impulsivity','cbqsf_activity','cbqsf_hip','cbqsf_approach']].mean(axis=1)
```

```python
data.to_csv('../scored_data/affect_scores_combined_20201217.csv')
```
